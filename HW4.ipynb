## Part 1

```python
# Import necessary libraries
import pandas as pd
```

```python
# Save CSV files to DataFrames
region1 = pd.read_csv('region1.csv')
region2 = pd.read_csv('region2.csv')
region3 = pd.read_csv('region3.csv')
supplier = pd.read_csv('supplier.csv')
```

```python
# View the first few rows of each DataFrame
region1.head()
region2.head()
region3.head()
```

### 1)

```python
# Concatenate the DataFrames
Store = pd.concat([region1, region2, region3], ignore_index=True)
```

```python
Store
```

### 2)

```python
# Replace missing values with "Not Reported"
Store.fillna("Not Reported", inplace=True)
```

```python
Store
```

### 3)

```python
# Seperate elements of Address column
Store[['Street', 'City', 'State', 'Zip']] = Store['Address'].str.split((','), expand=True)
```

```python
Store
```

```python
# Remove Address column
Store.drop(columns=['Address'], inplace=True)
```

```python
Store
```

### 4)

```python
# Pivot Quarter columns into a single column
Store = Store.melt(id_vars=['Store_code', 'Street', 'City', 'State', 'Zip'], 
                   value_vars=['Q1', 'Q2', 'Q3', 'Q4'],
                   var_name='Quarter', value_name='Sales')
```

```python
Store
```

### 5)

```python
# Merge Store with supplier DataFrame using an inner join between Store['State'] and supplier['Supp_State']
Store = Store.merge(supplier, left_on='State', right_on='Supp_State', how='inner')
```

### 1)

```python
billboard = pd.read_csv('billboard.csv')
```

```python
billboard.head()
```

```python
billboard.info()
```

#### a.

```python
# Define function to count the number of weeks with non-null values
def bill_week(row):
    week_cols = [f'wk{i}' for i in range(1, 77)]
    return row[week_cols].notnull().sum()
```

#### b.

```python
# Use the apply method to create a new column 'weeks_on_bill'
billboard['weeks_on_bill'] = billboard.apply(bill_week, axis=1)
billboard[['artist', 'track', 'weeks_on_bill']]
```

### 2)

```python
# Group by artist to get the number of tracks, and the min and max dates a song entered the billboard
artist_summary = billboard.groupby('artist').agg(
    num_tracks=('track', 'count'),
    first_date=('date.entered', 'min'),
    last_date=('date.entered', 'max')
).reset_index()

# Convert datetime stamps to dates only
artist_summary['first_date'] = pd.to_datetime(artist_summary['first_date']).dt.date
artist_summary['last_date'] = pd.to_datetime(artist_summary['last_date']).dt.date
```
